```python
models = {
    'RandomModel': {'RandomModel'},
    'SimpleCPUThresholdModel': {'SimpleCPUThresholdModel'},
    'DQN': {'MlpPolicy', 'CnnPolicy'},
    'PPO': {'MlpPolicy', 'CnnPolicy'},
    'RecurrentPPO': {'MlpLstmPolicy', 'CnnLstmPolicy'},
}
```

```python
# NEW
for model, policies in models.items():
    for policy in policies:
        predictions = np.squeeze(np.load(f'{BASE_PATH2}/actions/actions_{model.lower()}_{policy}.npy'), axis=1)
        N = 1000
        idxs = np.random.choice(np.arange(predictions.shape[0]), size=N)
        predictions = predictions[idxs]
        if len(policies) == 1:
            title = f'Action histogram for {model}'
        else:
            title = f'Action histogram for {model} with {policy} policy.'
        plot_action_histogram(predictions, title=title)
```
