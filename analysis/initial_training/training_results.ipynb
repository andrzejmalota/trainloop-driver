```python
models = {
    'DQN': {'MlpPolicy', 'CnnPolicy'},
    'PPO': {'MlpPolicy', 'CnnPolicy'},
    'RecurrentPPO': {'MlpLstmPolicy', 'CnnLstmPolicy'},
}

BASE_PATH = '../../initial_model'
```

## REWARDS DURING TRAINING

```python
for algo, policies in models.items():
    for policy in policies:
        try:
            plot_training_data(f"{BASE_PATH}/{algo.lower()}/{policy}/training_data.csv", title=f'{algo} : {policy}', data_type='reward')
        except FileNotFoundError:
            continue
            plot_training_data(f"{BASE_PATH}/{algo.lower()}/{policy}/best_model_rewards.csv", title=f'{algo} : {policy}', data_type='reward')
```
